## Task 1 -  GridSampler and TPESampler

```python
import pandas as pd
import matplotlib.pyplot as plt

study_grid = pd.read_csv("results_grid_sampler.csv")
study_tpe = pd.read_csv("results_tpe_sampler.csv")
trials = study_grid.number
values_grid = study_grid.value
values_tpe = study_tpe.value
sorted_values_tpe = values_tpe.sort_values()
sorted_values_grid = values_grid.sort_values()
max_values_grid = values_grid.cummax()
max_values_tpe = values_tpe.cummax()


fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plot without sorted values
ax[1].plot(trials, sorted_values_grid, label="Grid Sampler")
ax[1].plot(trials, sorted_values_tpe, label="TPE Sampler")
ax[1].legend()
ax[1].set_xlabel("#Trials")
ax[1].set_ylabel("Best Accuracy")
ax[1].grid()
ax[1].set_title("Sorted max")

# Plot with sorted values
ax[0].plot(trials, max_values_grid, label="Grid Sampler")
ax[0].plot(trials, max_values_tpe, label="TPE Sampler")
ax[0].legend()
ax[0].set_xlabel("#Trials")
ax[0].set_ylabel("Best Accuracy")
ax[0].grid()
ax[0].set_title("Cummulative max")

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
study_compression = pd.read_csv("results_compression.csv")
trials = study_compression.number
values = study_compression.value
max_values = values.cummax()
max_values_sorted = values.sort_values().cummax()

ax[0].plot(trials, max_values)
ax[0].set_xlabel("#Trials")
ax[0].set_ylabel("Best Accuracy")
ax[0].grid()
ax[0].set_title("Cummulative max")

ax[1].plot(trials, max_values_sorted)
ax[1].set_xlabel("#Trials")
ax[1].set_ylabel("Best Accuracy")
ax[1].set_title("Sorted max")
ax[1].grid()
plt.show()
```
