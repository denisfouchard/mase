## Task 1 -  GridSampler and TPESampler

```python
import pandas as pd
import matplotlib.pyplot as plt
study_grid = pd.read_csv("results_grid_sampler.csv")
study_tpe = pd.read_csv("results_tpe_sampler.csv")
trials = study_grid.number
values_grid = study_grid.value
values_tpe = study_tpe.value
max_values_grid = values_grid.sort_values().cummax()
max_values_tpe = values_tpe.sort_values().cummax()

plt.plot(trials, max_values_grid, label="Grid Sampler")
plt.plot(trials, max_values_tpe, label="TPE Sampler")
plt.legend()
plt.xlabel("#Trials")
plt.ylabel("Best Accuracy")
plt.grid()
plt.show()
```
